## Logistic Regression with regularization

We implement a regularized logistic regression model on example data (​ex2data2.mat​)

We vectorize the cost function:
> $J(θ) = \frac{1}{m}$$\sum_{i=1}^{m} [-y^i log( h_{θ} (x^i)) - (1-y^i)log(1-h_{θ}(x^i)) ] + \frac {\lambda} {2\mu} \sum_{j=1}^{n} θ^{2}_{j} $

- Generate the decision boundary of the example data via our logistic regression model.
- Plot the output boundary and the input data in the same figure.
- Explain the difference between logistic regression and regularized logistic regression and why regularization can improve the performance.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%config InlineBackend.figure_format = 'retina'
plt.style.use('seaborn-darkgrid')
```

### Loading the dataset

```python
X_feat = []
y_target = []

with open('data/ex2data2.mat') as ex2_d2:
    for line in ex2_d2:
        line = line.strip().split(',')
        X_feat.append([float(line[0]), float(line[1])])
        y_target.append(float(line[2]))

X_feat = np.array(X_feat)
y_target = np.array(y_target)
```

### Visualizing the dataset

```python
y_cmap = ['g' if i else 'r' for i in y_target]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=X_feat[:,0], ys=X_feat[:,1], zs=y_target, c=y_cmap)
plt.xlabel("Parameter 1 value")
plt.ylabel("Parameter 2 value")
plt.title("Green = 1 and Red = 0")
plt.show()
```

```python
plt.scatter(X_feat[:,0], X_feat[:,1], c=y_cmap)
plt.xlabel("Parameter 1 value")
plt.ylabel("Parameter 2 value")
plt.title("Green = 1 and Red = 0")
plt.show()
```
